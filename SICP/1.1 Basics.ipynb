#### Exercise 3: Defining a simple procedure

The number of combinations of $n$ things taken $k$ at a time is given by
$n!/k!(n-k)!$. Define a procedure `(comb n k)` that computes the number
of combinations, and find the number of combinations of 243 things taken
90 at a time. 

````scheme
(define (comb n k)
  (/ (fact n) (* (fact k) (fact (- n k)))))
````

````scheme
(comb 243 90)
````
